```python

import tkinter as tk
import random

# Function to decide the game winner
def decide_winner(user_choice, computer_choice):
    if user_choice == computer_choice:
        return "It's a tie!", "gray"
    elif (user_choice == "rock" and computer_choice == "scissors") or \
         (user_choice == "paper" and computer_choice == "rock") or \
         (user_choice == "scissors" and computer_choice == "paper"):
        return "You win!", "green"
    else:
        return "Computer wins!", "red"

# Function to manage user's selection and determine the game's outcome
def user_select(choice):
    global user_score, computer_score
    options = ["rock", "paper", "scissors"]
    computer_choice = random.choice(options)
    result, color = decide_winner(choice, computer_choice)
    result_label.config(text="Computer's Choice: " + computer_choice + "\n" + result, fg=color)
    if result == "You win!":
        user_score += 1
    elif result == "Computer wins!":
        computer_score += 1
    update_score()

# Function to update the scoreboard
def update_score():
    score_label.config(text=f"User: {user_score}  Computer: {computer_score}")

# Function to reset the game
def reset_game():
    global user_score, computer_score
    user_score = 0
    computer_score = 0
    update_score()
    result_label.config(text="")

# Create the main application window
win = tk.Tk()
win.title("Rock Paper Scissors Game")
win.configure(bg='pink')

# Buttons for user's choice
rock_btn = tk.Button(win, text="Rock", command=lambda: user_select("rock"),bg="grey")
rock_btn.pack()

paper_btn = tk.Button(win, text="Paper", command=lambda: user_select("paper"),bg="yellow")
paper_btn.pack()

scissors_btn = tk.Button(win, text="Scissors", command=lambda: user_select("scissors"),bg="blue")
scissors_btn.pack()

# Label to display the game result
result_label = tk.Label(win, text="",font=('Helvetica', 14, 'bold'))
result_label.pack()

# Scoreboard
user_score = 0
computer_score = 0
score_label = tk.Label(win, text="", font=('Helvetica', 12))
score_label.pack()
update_score()

# Play Again button
play_again_btn = tk.Button(win, text="Play Again", command=reset_game, bg="orange")
play_again_btn.pack()

# Launch the main event loop
win.mainloop()

```
