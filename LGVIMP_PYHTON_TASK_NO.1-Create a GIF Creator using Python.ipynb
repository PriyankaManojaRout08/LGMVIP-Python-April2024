```python
from moviepy.editor import *
from moviepy.config import change_settings

IMAGEMAGICK_PATH = r"C:\Program Files\ImageMagick-7.1.1-Q16-HDRI\magick.exe"
change_settings({"IMAGEMAGICK_BINARY": IMAGEMAGICK_PATH})
 
# Creating a ImageSequenceClip with the list of images 
clip = ImageSequenceClip(["pair_age.png","correlation_matrix.png","nationality_pie.png","gender_count.png",
                          "boxplot_age.png","heat_map.png","countplot_gender.png","scatter_ab_colored_sattus.png",
                          "hist_age bracket.png","line_cs.png"], fps=0.5) 
  
# Create a TextClip with the desired text and position
text_clip = TextClip("Some charts through which we can visualize data graphically!",
                     font ="Arial", fontsize=14, color='black')
text_clip = text_clip.set_position("center", "bottom").set_duration(5)

# Overlay the TextClip on top of the ImageSequenceClip
final_clip = CompositeVideoClip([clip, text_clip])

# Write the final clip to a GIF file
final_clip.write_gif("output.gif",fps=0.5)
final_clip.ipython_display(width = 360) 
```
